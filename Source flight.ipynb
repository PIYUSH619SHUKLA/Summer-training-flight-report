# Data Visualization

## Airline wise analysis

```python
data_airline_price = df.groupby(['airline'], as_index=False)['price'].mean()
data_airline_count = df.groupby(['airline'], as_index=False)['flight'].count()
print(data_airline_price)
print(data_airline_count)
```

```python
#sort in descending order
air_dec_count = data_airline_count.sort_values(by='flight', ascending=False)
air_dec_price = data_airline_price.sort_values(by='price', ascending=False)

# plots
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

#countplot
sns.countplot(x='airline', data=df, ax=ax[0],order=air_dec_count['airline'])
ax[0].set_title('Flight Count - Airline')

# barplot(average price)
sns.barplot(x='airline', y='price', data=data_airline_price, ax=ax[1],order=air_dec_price['airline'])
ax[1].set_title('Average Price - Airline')


plt.show()
```

## Conclusion:
### Vistara airlines has most flights but has highest price tickets while AirAsia airlines have lowest price tickets but have less number of flights

## Duration Wise Analysis

```python
data_duration_count= df.groupby(['duration'], as_index=False)['flight'].count()
data_duration_price = df.groupby(['duration'], as_index=False)['price'].mean()

print(data_duration_count)
print(data_duration_price)
```

```python
# Sort top 5 durations in descending order
top_5_duration_count = data_duration_count.sort_values(by='flight', ascending=False).head(5)
top_5_duration_price = data_duration_price.sort_values(by='price', ascending=False).head(5)

#subset dataframe
top_5_duration_subset = df[df['duration'].isin(top_5_duration_count['duration'])]

# plot
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

# countplot
sns.countplot(x='duration',data=top_5_duration_subset, ax=ax[0], order=top_5_duration_count['duration'])
ax[0].set_title('Flight Count - Top 5 Duration')


# barplot(average Price)
sns.barplot(x='duration', y='price', data=top_5_duration_price, ax=ax[1],order=top_5_duration_price['duration'])
ax[1].set_title('Average Price - Top 5 Duration')


plt.show()
```

## Conclusion:
### Number of flights having duration two hours and seventeen minutes are more and average ticket price are highest for flights having duration eighteen hours and forty-two minute.
